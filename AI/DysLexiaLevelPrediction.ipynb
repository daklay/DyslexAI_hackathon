```python
import random
import pandas as pd
import numpy as np

#def classify_based_on_percent(percent):

data = {"X": np.array([]), "Y": []}

for _ in range(1000):
    questions = random.randint(1, 100)  # random number of questions between 1 and 100
    correct_answers = random.randint(0, questions)  # random correct answers cannot exceed the number of questions
    ratio = correct_answers / questions * 100  # correct answers ratio as a percentage
    data["X"] = np.append(data["X"], ratio)

    if ratio <= 20:
        data["Y"].append('L1')
    elif 20 < ratio < 80:
        data["Y"].append('L2')
    else:
        data["Y"].append('L3')

df = pd.DataFrame(data)
#df['X']= df['X']100
```

```python

df
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# X contains Correct answers ratio percentages and y contains corresponding class labels
X1 = df['X'].array.reshape(-1, 1)  # we have to reshape because sklearn expects 2d array for features
y1 = df['Y']
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
# Train the classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X1, y1)

# Let's predict class label based on percentage
class_label = clf.predict(X_test)

# Compute metrics
accuracy = accuracy_score(y_test, class_label)
precision = precision_score(y_test, class_label, average='weighted')  # use 'weighted' for multi-class problem
recall = recall_score(y_test, class_label, average='weighted')
f1 = f1_score(y_test, class_label, average='weighted')
print('Accuracy: ', accuracy)
print('Precision: ', precision)
print('Recall: ', recall)
print('F1 Score: ', f1)

print('\nClassification Report\n')
print(classification_report(y_test, class_label))

print('\nConfusion Matrix\n')
print(confusion_matrix(y_test, class_label))
print(class_label)  # it would print 'L2', 'L1' or 'L3' based on trained model

```
